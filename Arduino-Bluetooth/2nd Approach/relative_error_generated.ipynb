```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
from sklearn.linear_model import LinearRegression
```

# Creating Pandas Dataframe from Relative error Reading Data

```python
df_05v = pd.read_csv('../docs/0.5v_1000data_250delay.csv')
df_05v = df_05v["voltage"]
df_3v = pd.read_csv('../docs/3v_1000data_250delay.csv')
df_3v = df_3v["voltage"]
df_6v = pd.read_csv('../docs/6v_1000data_250delay.csv')
df_6v = df_6v["voltage"]
df_9v = pd.read_csv('../docs/9v_1000data_250delay.csv')
df_9v = df_9v["voltage"]
df_12v = pd.read_csv('../docs/12v_1000data_250delay.csv')
df_12v = df_12v["voltage"]
df_15v = pd.read_csv('../docs/15v_1000data_250delay.csv')
df_15v = df_15v["voltage"]
df_18v = pd.read_csv('../docs/18v_1000data_250delay.csv')
df_18v = df_18v["voltage"]
df_21v = pd.read_csv('../docs/21v_1000data_250delay.csv')
df_21v = df_21v["voltage"]
df_24_5v = pd.read_csv('../docs/24.5v_1000data_250delay.csv')
df_24_5v = df_24_5v["voltage"]
df = {
    "0.5v" : [],
    "3v" : [],
    "6v" : [],
    "9v" : [],
    "12v" : [],
    "15v" : [],
    "18v" : [],
    "21v" : [],
    "24.5v" : []
}
for i in df_05v:
    df["0.5v"].append(((i-0.5)/0.5) * 100)
for i in df_3v:
    df["3v"].append(((i-3)/3) * 100)
for i in df_6v:
    df["6v"].append(((i-6)/6) * 100)
for i in df_9v:
    df["9v"].append(((i-9)/9) * 100)
for i in df_12v:
    df["12v"].append(((i-12)/12) * 100)
for i in df_15v:
    df["15v"].append(((i-15)/15) * 100)
for i in df_18v:
    df["18v"].append(((i-18)/18) * 100)
for i in df_21v:
    df["21v"].append(((i-21)/21) * 100)
for i in df_24_5v:
    df["24.5v"].append(((i-24.5)/24.5) * 100)

df = pd.DataFrame(df)
df = df.apply(pd.to_numeric)
df = df.apply(lambda x: x.sort_values().values)
df.to_csv('../docs 2nd Approach/all_voltage.csv', index=False)
df
```

# Generated Z Score from each relative error

```python
for i in df.columns:
    df[i] = (df[i] - df[i].mean()) / df[i].std()

df.to_csv('../docs 2nd Approach/z-score_all_voltage.csv', index=False)
df
```
