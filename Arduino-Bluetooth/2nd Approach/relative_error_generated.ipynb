```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
from sklearn.linear_model import LinearRegression
```

# Creating Pandas Dataframe from Relative error Reading Data

```python
df_05v = pd.read_csv('../capture3rd/0.5v.csv')
df_05v = df_05v["voltage"]
df_3v = pd.read_csv('../capture3rd/3v.csv')
df_3v = df_3v["voltage"]
df_6v = pd.read_csv('../capture3rd/6v.csv')
df_6v = df_6v["voltage"]
df_9v = pd.read_csv('../capture3rd/9v.csv')
df_9v = df_9v["voltage"]
df_12v = pd.read_csv('../capture3rd/12v.csv')
df_12v = df_12v["voltage"]
df_15v = pd.read_csv('../capture3rd/15v.csv')
df_15v = df_15v["voltage"]
df_18v = pd.read_csv('../capture3rd/18v.csv')
df_18v = df_18v["voltage"]
df_21v = pd.read_csv('../capture3rd/21v.csv')
df_21v = df_21v["voltage"]
df_24_5v = pd.read_csv('../capture3rd/24,5v.csv')
df_24_5v = df_24_5v["voltage"]

df = {
    "0.5v" : df_05v.to_list(),
    "3v" : df_3v.to_list(),
    "6v" : df_6v.to_list(),
    "9v" : df_9v.to_list(),
    "12v" : df_12v.to_list(),
    "15v" : df_15v.to_list(),
    "18v" : df_18v.to_list(),
    "21v" : df_21v.to_list(),
    "24.5v" : df_24_5v.to_list()
}

df = pd.DataFrame(df)
df = df.apply(pd.to_numeric)
df = df.apply(lambda x: x.sort_values().values)
df.to_csv('../docs 2nd Approach/all_voltage.csv', index=False)
df
```

```python

df = {
    "0.5v" : [],
    "3v" : [],
    "6v" : [],
    "9v" : [],
    "12v" : [],
    "15v" : [],
    "18v" : [],
    "21v" : [],
    "24.5v" : []
}
for i in df_05v:
    df["0.5v"].append(((i-0.5)/0.5) * 100)
for i in df_3v:
    df["3v"].append(((i-3.0)/3.0) * 100)
for i in df_6v:
    df["6v"].append(((i-6)/6) * 100)
for i in df_9v:
    df["9v"].append(((i-9)/9) * 100)
for i in df_12v:
    df["12v"].append(((i-12)/12) * 100)
for i in df_15v:
    df["15v"].append(((i-15)/15) * 100)
for i in df_18v:
    df["18v"].append(((i-18)/18) * 100)
for i in df_21v:
    df["21v"].append(((i-21.0)/21.0) * 100)
for i in df_24_5v:
    df["24.5v"].append(((i-24.5)/24.5) * 100)

df = pd.DataFrame(df)
df = df.apply(pd.to_numeric)
df = df.apply(lambda x: x.sort_values().values)
df.to_csv('../docs 2nd Approach/relative_error.csv', index=False)
df
```

# Generated Z Score from each relative error

```python
for i in df.columns:
    df[i] = (df[i] - df[i].mean()) / df[i].std()

df.to_csv('../docs 2nd Approach/z-score_all_voltage.csv', index=False)
df
```
