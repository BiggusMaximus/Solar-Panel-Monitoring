```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
```

```python
df = pd.read_csv('../docs 2nd Approach/remove_outlier_chauvenet.csv')
df
```

```python
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(8)


plt.plot(df.index, df['value'].std()/np.sqrt(df.index))
    
plt.xlabel("Sample Sizes")
plt.ylabel("Voltage Error (%)")
plt.legend()
plt.grid()
plt.savefig('../docs/error_std_width.png')
plt.show()
```
