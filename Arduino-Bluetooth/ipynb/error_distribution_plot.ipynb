```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../docs/error_voltage_all.csv')
df
```

```python
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(8)

for i in df.columns:
    plt.plot(df.index, df[i].std()/np.sqrt(df.index), label=i)
    
plt.axhline(y=1, linestyle="--", label="1%")
plt.xlabel("Sample Sizes")
plt.ylabel("Voltage Error (%)")
plt.legend()
plt.grid()
plt.savefig('../docs/error_std_width.png')
plt.show()
```
