```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy
import pandas as pd
import os
```

```python
df_05v = pd.read_csv('./docs/0.5v_1000data_250delay.csv')
df_05v = df_05v["voltage"]
df_3v = pd.read_csv('./docs/3v_1000data_250delay.csv')
df_3v = df_3v["voltage"]
df_6v = pd.read_csv('./docs/6v_1000data_250delay.csv')
df_6v = df_6v["voltage"]
df_9v = pd.read_csv('./docs/9v_1000data_250delay.csv')
df_9v = df_9v["voltage"]
df_12v = pd.read_csv('./docs/12v_1000data_250delay.csv')
df_12v = df_12v["voltage"]
df_15v = pd.read_csv('./docs/15v_1000data_250delay.csv')
df_15v = df_15v["voltage"]
df_18v = pd.read_csv('./docs/18v_1000data_250delay.csv')
df_18v = df_18v["voltage"]
df_21v = pd.read_csv('./docs/21v_1000data_250delay.csv')
df_21v = df_21v["voltage"]
```

```python
std_05v = df_05v.std()
std_3v = df_3v.std()
std_6v = df_6v.std()
std_9v = df_9v.std()
std_12v = df_12v.std()
std_15v = df_15v.std()
std_18v = df_18v.std()
std_21v = df_21v.std()
mean_05v = df_05v.mean()
mean_3v = df_3v.mean()
mean_5v = df_6v.mean()
mean_9v = df_9v.mean()
mean_12v = df_12v.mean()
mean_18v = df_18v.mean()
mean_21v = df_21v.mean()
```

```python
x = (std_05v/0.01)**2
x
```

```python
fig, axs = plt.subplots(4, 2, figsize=(8,15))
plt.tight_layout()
x = np.arange(0, 1000, 1)
axs[0, 0].plot(x, std_05v/np.sqrt(x))
axs[0, 0].set_title('0.5v')
axs[0, 1].plot(x, std_3v/np.sqrt(x), 'tab:orange')
axs[0, 1].set_title('3v')
axs[1, 0].plot(x, std_6v/np.sqrt(x), 'tab:green')
axs[1, 0].set_title('6v')
axs[1, 1].plot(x, std_9v/np.sqrt(x), 'tab:red')
axs[1, 1].set_title('9v')
axs[2, 0].plot(x, std_12v/np.sqrt(x), 'tab:red')
axs[2, 0].set_title('12v')
axs[2, 1].plot(x, std_15v/np.sqrt(x), 'tab:red')
axs[2, 1].set_title('15v')
axs[3, 0].plot(x, std_21v/np.sqrt(x), 'tab:red')
axs[3, 0].set_title('18v')
axs[3, 1].plot(x, std_21v/np.sqrt(x), 'tab:red')
axs[3, 1].set_title('21v')
```

```python
f = plt.figure()
f.set_figwidth(5)
f.set_figheight(5)
plt.plot(x, 100 * std_05v/np.sqrt(x), label="0.5v", color='red')
plt.plot(x, 100 *std_3v/np.sqrt(x), label="3v", color='green')
plt.plot(x, 100 *std_6v/np.sqrt(x), label="6v", color='black')
plt.plot(x, 100 *std_9v/np.sqrt(x), label="9v", color='blue')
plt.plot(x, 100 *std_12v/np.sqrt(x), label="12v", color='magenta')
plt.plot(x, 100 *std_15v/np.sqrt(x), label="15v", color='yellow')
plt.plot(x, 100 *std_18v/np.sqrt(x), label="18v", color='cyan')
plt.plot(x, 100 *std_21v/np.sqrt(x), label="21v", color='red')
plt.axhline(y=1, linestyle="--", label="1%")
plt.xlabel("Sample Sizes")
plt.ylabel("Voltage Error (%)")
plt.legend()
plt.grid()
plt.savefig('../docs/error_std_width.png')
plt.show()
```
