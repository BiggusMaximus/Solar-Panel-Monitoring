```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
import matplotlib.dates
values = {
    "time" : [],
    "voltage" : [],
    "current" : []
}
n = 5
with open('bluetooth.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        # if n%5 == 0:
        values['time'].append(line.split()[0])
        values['voltage'].append(float(line.split()[2]))
        values['current'].append(float(line.split()[4]))
        print(n)
        #     n = n + 1
        # else:
        #     next
        #     n = n + 1

print(values)
df = pd.DataFrame(values, columns=['time','voltage', 'current'])
df['power'] = df['voltage'] * df['current']
```

```python
df.to_csv('data.csv')
df.to_excel('data.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as md
import dateutil
import matplotlib.ticker as plticker
loc = plticker.MultipleLocator(base=1.0) # this locator puts ticks at regular intervals

datestrings = df['time']
dates = [dateutil.parser.parse(s) for s in datestrings]

fig, ax = plt.subplots()
ax.set_xticks(dates)

xfmt = md.DateFormatter('%H:%M:%S')
print(ax.xaxis.set_major_formatter(xfmt))
ax.xaxis.set_major_formatter(xfmt)
ax.yaxis.set_major_locator(MaxNLocator(prune='lower'))

ax.plot(dates,df['current'], "o-")


fig.autofmt_xdate()
plt.xticks(rotation=45)
plt.show()
```
